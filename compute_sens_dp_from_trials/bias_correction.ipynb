```python
def get_data(indir, ns_str_list, var='ns'):
    
    data = []
    means = []

    for ns in ns_str_list:
        print(ns)
        vals = []
        inf = os.path.join(indir,  "mean_ns"+ns+"_trials20_"+"rss*.npy")
        print(inf)
        infs = sorted(glob.glob(inf))
        #print(infs)
        for inf in infs:
            dat = np.load(inf)
            vals += dat[var].tolist()

        data.append(np.percentile(vals, [2.5, 16,50,84, 97.5]))
        means.append(np.mean(vals))
        #print(data)
    return(data, means)
```
